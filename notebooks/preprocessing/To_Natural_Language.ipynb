```python
def to_paper(row):
  return f"""PAPER ID : {row['paper_id']}

  Title : {row['title']}

  Authors : {row['authors']}

  Abstract : {row['abstract']}

  Primary Category : {row['primary_category']}

  All Categories : {row['all_categories']}"""
```

```python
to_paper(df.loc[1])
```

```python
df['document'] = df.apply(to_paper, axis=1)
```

```python
df.sample(1)
```

```python
df.loc[208,'document']
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("BAAI/bge-small-en-v1.5")

embedding = model.encode(df.loc[208,'document'])
```

```python
embedding
```

```python
df['embedding'] = df['document'].apply(lambda doc : model.encode(doc))
```

```python
df[['document', 'embedding']].to_csv('Natural_Language_v.csv', index = False)
```
